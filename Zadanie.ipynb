```python
from typing import List
import random as rnd
import pandas as pd
import numpy as np

Matrix = List[List[float]]

def array_generator(n: int) -> Matrix:
    A = []
    for i in range(n):
        # список с возможными размерами массивов 
        sizes = list(range(1, 100))
        # мешаем список для "случайности"
        rnd.shuffle(sizes)
        # наш случайный размер для текущей итерации
        tempSize = sizes[i]
        # заполняем случайными числами текущий массив
        tempArray = [rnd.uniform(-1000, 1000) for _ in range(tempSize)]
        # выбираем как сортировать
        tempArray.sort() if i % 2 == 0 else tempArray.sort(reverse=True)
        A.append(tempArray)
                               
    return A

# тест
n = 5
A = array_generator(n)

for i in range(n):
    print("Массив #", i, ": ", A[i],"\n", "Размер:", len(A[i]), "\n")
```
